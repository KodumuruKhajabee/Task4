```python
from moviepy.editor import VideoFileClip

def chunk_video(video_path, chunk_duration=60, overlap=10):
    video = VideoFileClip(video_path)
    duration = int(video.duration)
    chunks = []

    start = 0
    index = 1

    while start < duration:
        end = min(start + chunk_duration, duration)

        subclip = video.subclip(start, end)
        output_name = f"shabbu_10th_chunk_{index}.mp4"

        subclip.write_videofile(
            output_name,
            codec="libx264",
            audio_codec="aac"
        )

        chunks.append(output_name)

        start = end - overlap
        index += 1

    video.close()
    return chunks


chunks = chunk_video(
    "shabbu 10th.mp4",
    chunk_duration=60,
    overlap=10
)

print("Generated chunks:", chunks)
```
